###Separate Input and Output

```python
# Drop 'Newspaper' column from the input
X = df.drop(columns=['Sales', 'Newspaper'])
y = df['Sales']
```

```python
X
```

```python
y
```

###Plot the original 3D scatter plot

```python
# Plot the original 3D scatter plot

fig = px.scatter_3d(df, x='TV', y='Radio', z='Sales')

fig.show()
```

###Train Linear Regression model

```python
# Step 4: Create and train Linear Regression model
model = LinearRegression()
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# Step 5: Take new input from user
tv = float(input("Enter TV advertisement budget: "))
radio = float(input("Enter Radio advertisement budget: "))

# Step 6: Create input DataFrame for prediction
new_data = pd.DataFrame({'TV': [tv],'Radio': [radio]})

# Step 7: Predict sales
predicted_sales = model.predict(new_data)

# Step 8: Display result
print(f"Predicted Sales: {predicted_sales[0]:.2f}")
```

```python
# Plot the original 3D scatter plot
fig = px.scatter_3d(df, x='TV', y='Radio', z='Sales')

# Add new predicted point (RED)
fig.add_scatter3d(x=[tv],y=[radio],z=[predicted_sales[0]],mode='markers',   marker=dict(size=8, color='red'),name='New Prediction')

fig.show()
```
