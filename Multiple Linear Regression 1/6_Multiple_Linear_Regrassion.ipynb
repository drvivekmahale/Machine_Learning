###Separate Input and Output

```python
# Drop 'Newspaper' column from the input
X = df.drop(columns=['Sales', 'Newspaper'])
y = df['Sales']
```

```python
X
```

```python
y
```

###Plot the 3D scatter plot

```python
# Plot the original 3D scatter plot

fig = px.scatter_3d(df, x='TV', y='Radio', z='Sales')

fig.show()
```

###Train Linear Regression model

```python
# Step 4: Create and train Linear Regression model
model = LinearRegression()
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# Create grid points for surface plot (use ranges based on TV and Radio)
tv_range = np.linspace(X['TV'].min(), X['TV'].max(), 10)
radio_range = np.linspace(X['Radio'].min(), X['Radio'].max(), 10)
tvGrid, radioGrid = np.meshgrid(tv_range, radio_range)
```

```python
# Create input data for prediction
final = np.vstack((tvGrid.ravel(), radioGrid.ravel())).T  # Shape: (100, 2)
```

```python
# Predict sales for grid points
z_final = model.predict(final).reshape(10, 10)
```

```python
# Create a 3D scatter plot
fig = px.scatter_3d(df, x='TV', y='Radio', z='Sales')
```

```python
# Add surface plot to visualize predicted sales
fig.add_trace(go.Surface(x=tv_range, y=radio_range, z=z_final, colorscale='Viridis', opacity=0.7))
# Show the plot
fig.show()
```
