## **Business Problem**

Develop a machine learning model capable of predicting the species of a penguin based on its physical attributes (such as bill length, flipper length, body mass, etc.). This model can assist field researchers in classifying penguins more efficiently without requiring the expertise of a biologist, thereby saving time and resources in wildlife studies.

# **Feature Engineering**

**Missing Value**

```python
# Check for missing values in each column
df.isna().sum()
```

```python
# Drop rows with any missing values
df = df.dropna()

# Display the new shape of the DataFrame
df.shape
```

# **Encoding**
**Why do we need it?**

Many machine learning models (like linear regression or neural networks) can’t work with text. They need numerical input. But if we just assign numbers (like 0, 1, 2), the model might think there's an order or relationship that doesn't actually exist.

```python
#pd.get_dummies(df.drop('species', axis=1), drop_first=True)
#X = pd.get_dummies(df.drop('species', axis=1), drop_first=True)
X = pd.get_dummies(df.drop('species', axis=1), drop_first=True, dtype=int)
X
```

```python

y = df['species']
```

**Train/Test Split**

```python
# Train/Test Split
from sklearn.model_selection import train_test_split

# Split the data: 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)


```

**Modelling**

```python
#Import the Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
#Create the Model
model = DecisionTreeClassifier()
```

```python
#Train (Fit) the Model
model.fit(X_train, y_train)
```

# **Prediction**

```python
base_pred = model.predict(X_test)

pred_train = model.predict(X_train)
```

# **Evaluation**

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, accuracy_score
```

```python
accuracy_score(y_train, pred_train)
```

```python
accuracy_score(y_test, base_pred)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
plt.show()
```

```python
print(classification_report(y_test, base_pred))
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(12,8), dpi=150)
plot_tree(model, filled=True, feature_names=X.columns)
plt.show()
```
